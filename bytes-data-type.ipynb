```python
type('abc')  # Sequence of characters
type(b'abc')  # Sequence of bytes
```

```python
data = b'abc'
len(data)
data[0]  # Each byte is an integer 0-255
data[1]
data[2]
```

```python
type(data[0])  # 1 Byte = 8 bits = (2^8=256) states (int 0-255)
```

```python
chr(data[0])  # Integer to character
chr(data[1])
chr(data[2])
```

```python
chr(35)  # Based on ASCII character set
chr(36)
chr(64)
chr(65)
```

```python
chr(255)  # Last char of extended ASCII
chr(20)  # Non printable character (0-31)
```

**Hexadecimal escape sequence**

\xHH

2 digits of hexadecimal (00 to FF)

Represents a Byte (16*16=256 states)

```python
'\x41'  # 65 in decimal
'\x42'  # 66
'\x43'  # 67
```

```python
type('\x41')
'\x41' == 'A'
```

```python
print('x41 is the first alphabet')
print('\x41 is the first alphabet')  # \xHH is a Byte
print('\\x41 is the first alphabet')  # \\ is literal '\'
```

```python
print('launch\\missile')  # \ is used to escape
print(r'launch\missile')  # Raw string
```

```python
data.decode()  # bytes -> str

''.join([chr(byte) for byte in data])  # same as this
```

**Format specifier in F strings**

```python
f'{97}'
f'{97:.2f}'
```

```python
f'{23412}'
f'{23412:,}'
f'{23412:,.2f}'
```

```python
f'{0:b}'  # Decimal to binary
f'{1:b}'
f'{2:b}'
f'{3:b}'
f'{4:b}'
f'{5:b}'
f'{22:b}'
```

```python
f'{0:b}'
f'{0:5b}'  # Decimal to binary with padding
f'{0:10b}'
```

```python
f'{3:b}'  # Exact binary without padding
f'{3:5b}'  # Padded with space
f'{3:05b}'  # Padded with zeros
```

```python
data = b'abc'
for i in data:
  print(i, chr(i), f'{i:08b}')  # Padding of 8
```

```python
list(b'abc')  # Bytes to array
bytes([97, 98, 99])  # Array to bytes
```

```python
for i in [97, 98, 99]:
  f'{i:08b}'
```

```python
''.join([f'{i:08b}' for i in data])  # Binary representation of entire data
' '.join([f'{i:08b}' for i in data])  # Bytes are space separated for visual clarity
```

```python
int('01100001 ', 2)  # Binary to decimal
chr(int('01100001 ', 2))  # Binary to decimal to character
```

```python
binary_str = '01100001 01100010 01100011'

for byte in binary_str.split():
  print(byte, int(byte, 2), chr(int(byte, 2)))
```
